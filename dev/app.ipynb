# Dev file for visualisation of the results

```python
import plotly.graph_objs as go
import elec_lca
from pathlib import Path
```

```python
df_scenarios = elec_lca.reading.read_user_input_template_excel_file('../data', 'user_input_template.xlsm')
```

```python
df_scenarios.head()
```

```python
layout = go.Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)

fig = go.Figure(layout=layout)
df_scenario = df_scenarios[df_scenarios.scenario == 'NZ50']
df_scenario = df_scenario.sort_values(by=['period'])
tech_list = list(df_scenario.technology.unique())
t = list(df_scenario.period.unique())

for tech in tech_list:
    data = list(100*df_scenario[df_scenario.technology == tech].value)
    fig.add_trace(go.Scatter(
        x=t, y=data,
        #hoverinfo='x+y',
        mode='lines',
        name=tech,
        #line=dict(width=0.5, color='rgb(131, 90, 241)'),
        stackgroup='one' # define stack group
    ))

fig.update_layout(
    xaxis_title="Years",
    yaxis_title="Shares [%]",
    legend_title="Technologies",
)

fig.update_traces(mode="markers+lines", hovertemplate=None)
fig.update_layout(hovermode="x unified")

fig.show()
```
